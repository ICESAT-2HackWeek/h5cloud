# Running xarray with an h5coro backend

This notebook shows the example syntax for opening a data structure in xarray with an MVP h5coro backend. In order to be run a dev version, xarray needs to be installed using the IS2 hackweek fork: [https://github.com/ICESAT-2HackWeek/xarray](https://github.com/ICESAT-2HackWeek/xarray). h5coro also needs to be installed [https://github.com/ICESat2-SlideRule/h5coro](https://github.com/ICESat2-SlideRule/h5coro).

The bounds of this MVP version are:
- a group argument is required
- s3 urls only (no local filepaths)
- multidimensional variables are dropped

The MVP code works on ALT03 `heights` data from any beam, but we hope to continue to generalize!

```python
import xarray as xr
import earthaccess
```

```python
auth = earthaccess.login()
s3url_atl03 = 'nsidc-cumulus-prod-protected/ATLAS/ATL03/006/2019/11/30/ATL03_20191130112041_09860505_006_01.h5'
ds = xr.open_dataset(s3url_atl03, engine='h5coro', group='/gt1l/heights', creds=auth)
ds
```
